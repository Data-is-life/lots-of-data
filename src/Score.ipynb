```python
import pandas as pd
import chess
import chess.pgn
```

```python
pgn = open('../data/dest.pgn')
game = chess.pgn.read_game(pgn)

board = game.board()

fen_moves = []

for move in game.main_line():
    board.push(move)
    fen_moves.append(board.fen().replace(
        ' w KQ -', '').replace(' b KQ -', '').replace(' w KQkq -','').replace(
        ' b KQkq -','').replace(' w kq -','').replace(' b kq -','').replace(' b -',''))

fen_moves
```

```python
bl = []
wh = []

for num in fen_moves:
    bl.append(
        num.count('p')+num.count('b')*3+num.count('r')*5+num.count(
            'n')*3+num.count('q')*9)
    wh.append(
        num.count('P')+num.count('R')*5+num.count(
            'B')*3+num.count('N')*3+num.count('Q')*9)
wh_score = [x - y for x, y in zip(wh, bl)]
bl_score = [x - y for x, y in zip(bl, wh)]
wh_score
bl_score
```
